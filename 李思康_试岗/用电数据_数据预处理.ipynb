```python
import pandas as pd
import numpy as np
```

### 1.数据读取和基础清洗

```python
df = pd.read_excel("./用电数据.xlsx",header=1)
df.head()
```

```python
df.columns = ['timestamp', 'active_energy', 'current_A', 'current_B', 'current_C', 
              'power_A', 'power_B', 'power_C', 'apparent_power']
```

```python
df.head()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
print(df.dtypes)
```

### 2.缺失值处理策略

```python
df.info()
```

```python
missing_ratio = df.isnull().mean()
missing_ratio
```

```python
df = df.dropna()
```

```python
df.info()
```

### 3. 异常值检测与修正

```python
df.describe()
```

```python
numeric_cols = [
    "A相电流", "B相电流", "C相电流",
    "A相有功功率", "B相有功功率", "C相有功功率",
    "视在功率"
]
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
original_rows = len(df)
cleaned_rows = len(df)
removed_rows = original_rows - cleaned_rows
print(f"处理前数据行数：{original_rows}")
print(f"处理后数据行数：{cleaned_rows}")
print(f"删除的异常值行数：{removed_rows}")
```

### 4.特征工程

```python
# 时间特征衍生
df['hour'] = df['timestamp'].dt.hour
df['is_workday'] = df['timestamp'].dt.dayofweek < 5

# 电气特征计算
df['total_power'] = df[['power_A','power_B','power_C']].sum(axis=1)
df['power_factor'] = df['total_power'] / df['apparent_power']  # 功率因数
df['load_imbalance'] = df[['current_A','current_B','current_C']].std(axis=1)  # 三相不平衡度

# 能耗增量计算
df['energy_delta'] = df['active_energy'].diff().fillna(0)
```

```python
df.head()
```

### 5.数据归一化

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaled_cols = ['current_A','current_B','current_C','power_A','power_B','power_C','apparent_power']
df[scaled_cols] = scaler.fit_transform(df[scaled_cols])
```

```python
df.head()
```

### 6.时间序列重采样

```python
df = df.set_index('timestamp').resample('1H').mean().reset_index()
df.head()
```

```python
df.to_excel("数据处理后的用电数据_可用于建模.xlsx")
```
